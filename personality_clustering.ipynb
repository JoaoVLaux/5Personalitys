```python
import pandas as pd
```

```python
# Carrengando dataset
df = pd.read_csv("data-final.csv", sep= '\t')
# Tirando o limte de colunas
pd.options.display.max_columns = None
pd.options.display.precision = 2
```

```python
# Primeira visualição no dataset
df.head()
```

```python
df.shape
```

```python
dropar = df.columns[50:111]
df.drop(columns=dropar, inplace=True)
```

```python
df.head()
```

```python
df.describe()
# Existem valores com numero 0 ! !  
```

```python
df[(df == 0).all(axis=1)]
```

```python
df = df[(df > 0).all(axis=1)]
```

```python
df.describe()
```

```python
# DF pronto para clusterizar
# Clusterizar utilizando KMeans do SKLEARN
from sklearn.cluster import KMeans
```

```python
 #N° de cluesters foi definido em 5 pela regra de negócio "5 personalidades"
kmeans = KMeans(n_clusters=5)
k_fit = kmeans.fit(df)
```

```python
k_personalidade = k_fit.labels_
df["Personalidades"] = k_personalidade
```

```python
df["Personalidades"].value_counts()
```

```python
# Criando valores para novo dataset
ext = df.columns[0:10]
est = df.columns[10:20]
agr = df.columns[20:30]
csn = df.columns[30:40]
opn = df.columns[40:50]
```

```python
# DataFrame com médias
new_data = pd.DataFrame()
new_data['Extroversão'] = df[ext].sum(axis=1)/10
new_data['Neurotico'] = df[est].sum(axis=1)/10
new_data['Agradável'] = df[agr].sum(axis=1)/10
new_data['Consciente'] = df[csn].sum(axis=1)/10
new_data['Aberto'] = df[opn].sum(axis=1)/10
new_data['Personalidades'] = k_personalidade
```

```python
new_data.head()
```

```python
clusters = new_data.groupby("Personalidades").mean()
```

```python
# Importanto MatPlotLib
import matplotlib.pyplot as plt
```

```python
# Plotando Grafico dos 5 Grupos
plt.figure(figsize=(30,6))
for i in range(0, 5):
    plt.subplot(1,5,i+1)
    plt.bar(clusters.columns, clusters.iloc[:, i], color='green', alpha=0.4)
    plt.plot(clusters.columns, clusters.iloc[:, i], color='red')
    plt.title('Grupo ' + str(i))
    plt.xticks()
    plt.ylim(0,4);
```

```python
from sklearn.cluster import SpectralBiclustering
df.shape
```

```python
spec = SpectralBiclustering(n_clusters=5)
spec_fit = spec.fit(df)
```

```python
spec_personalidade = spec_fit.row_labels_
```

```python
df["Personalidades Spec"] = spec_personalidade
```

```python
df["Personalidades Spec"].value_counts()
```

```python
# DataFrame com médias para o Spec
spec_data = pd.DataFrame()
spec_data['Extroversão'] = df[ext].sum(axis=1)/10
spec_data['Neurotico'] = df[est].sum(axis=1)/10
spec_data['Agradável'] = df[agr].sum(axis=1)/10
spec_data['Consciente'] = df[csn].sum(axis=1)/10
spec_data['Aberto'] = df[opn].sum(axis=1)/10
spec_data['Personalidades'] = spec_personalidade
```

```python
clusters_spec = spec_data.groupby("Personalidades").mean()
```

```python
plt.figure(figsize=(30,6))
for i in range(0, 5):
    plt.subplot(1,5,i+1)
    plt.bar(clusters_spec.columns, clusters_spec.iloc[:, i], color='green', alpha=0.4)
    plt.plot(clusters_spec.columns, clusters_spec.iloc[:, i], color='red')
    plt.title('Grupo ' + str(i))
    plt.xticks()
    plt.ylim(0,4);
```
